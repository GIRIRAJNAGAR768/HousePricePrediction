```python
y=train.SalePrice
train.drop({"SalePrice"},axis=1,inplace=True)
```

```python
all_data = train.append(test)
```

```python
all_data.drop({"Id","Fence","Alley","PoolQC","MiscFeature","FireplaceQu","Utilities"},axis=1,inplace=True)
```

```python
all_data["TotalFA"] = all_data["TotalBsmtSF"]+all_data["1stFlrSF"]+all_data["2ndFlrSF"]
all_data.drop({"TotalBsmtSF","1stFlrSF","2ndFlrSF"},axis=1,inplace=True)
```

```python
all_data.info()
```

```python
cat = all_data.select_dtypes(include=['object']).columns
name=[]
for i in cat:
    name.append(i)
```

```python
for j in name:
    all_data[j] = all_data[j].fillna(all_data[j].mode()[0])
```

```python
flot = all_data.select_dtypes(include=['float64']).columns
name2=[]
for i in flot:
    name2.append(i)
```

```python
for j in name2:
    all_data[j] = all_data[j].fillna(all_data[j].median())
```

```python
inti = all_data.select_dtypes(include=['int64']).columns
name3=[]
for i in inti:
    name3.append(i)
```

```python
non_cat=all_data.drop(name,axis=1)
```

```python
cat_data=all_data.drop(non_cat,axis=1)
```

```python
dummy=pd.get_dummies(cat_data,drop_first=True)
```

```python
frame=[dummy,non_cat]
Final=pd.concat(frame,axis=1)
```

```python
train=Final.iloc[:1436,:]
```

```python
test=Final.iloc[1436:,:]
```

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=100)
```

```python
reg.fit(train,y)
```

```python
reg.score(train,y)
```

```python
y_pred=reg.predict(test)
```
